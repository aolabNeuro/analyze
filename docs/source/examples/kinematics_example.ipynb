```python
import os
import numpy as np
import aopy
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('paper')
```

```python
base_dir = ""
preprocessed_filename = 'preprocessed_te1927.hdf'
data = aopy.data.load_hdf_group(base_dir, preprocessed_filename, 'exp_data')
metadata = aopy.data.load_hdf_group(base_dir, preprocessed_filename, 'exp_metadata')
cursor_kinematics = data['task']['cursor'] # cursor (x, z, y) position on each bmi3d cycle
hand_kinematics = data['task']['manual_input'] # hand kinematics (x, y, z) in Optitrack coordinates received by bmi3d before transformation to 2D
time = data['clock']['timestamp']
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(time, cursor_kinematics[:, 0])
ax[0].set_ylabel('X pos (cm)')
ax[1].plot(time, cursor_kinematics[:, 2])
ax[1].set_ylabel('Y pos (cm)')
```

```python
# the raw optitrack data can be a bit noisy
hand_kinematics[hand_kinematics > 1000] = np.nan
hand_kinematics[hand_kinematics == 0] = np.nan
```

```python
fig, ax = plt.subplots(3,1)
ax[0].plot(time, hand_kinematics[:, 0])
ax[0].set_ylabel('X pos (cm)')
ax[1].plot(time, hand_kinematics[:, 1])
ax[1].set_ylabel('Y pos (cm)')
ax[2].plot(time, hand_kinematics[:, 2])
ax[2].set_ylabel('Z pos (cm)')
```
