```python
def update_graph(num):
    # data=dat[dat['time']==num]
    data=dat.loc[num+11000,:]
    graph1.set_data (data.x, data.y)
    graph1.set_3d_properties(data.z)

    graph2.set_data (data.x2, data.y2)
    graph2.set_3d_properties(data.z2)

    graph3.set_data (data.x3, data.y3)
    graph3.set_3d_properties(data.z3)
    title.set_text('3D Test, frame={}'.format(num))
    #print(data)
    return title, graph, 


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
title = ax.set_title('3D Test')

data=dat[dat['time']==0]
graph1, = ax.plot(data.x, data.y, data.z, linestyle="solid", marker="o")
graph2, = ax.plot(data.x2, data.y2, data.z2, linestyle="solid", marker="o")
graph3, = ax.plot(data.x3, data.y3, data.z3, linestyle="solid", marker="o")

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-0.2,0.2)
ax.set_ylim(0.6,1)
ax.set_zlim(0.1,0.5)

ani = matplotlib.animation.FuncAnimation(fig, update_graph, 20000 , interval = 1 ,blit=True)

ani.save('Rigidbody_movement.mp4')
plt.show() 
```
